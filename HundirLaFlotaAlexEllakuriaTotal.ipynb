```python
from variablesHLF import *
from funcionesHLF import *
import random
import sys


lstTableroMac = []
contDisparos = 0

lstTableroMac = crearTablero(lstTableroMac)
lstTableroMac = ponerTodosBarcoMac(lstTableroMac)

pintarTablero(lstTableroMac)

bolSeguir = True
try:
    while bolSeguir:
        print("\nIntroduce coordenadas (1 a", TamañoTablero, ")")
        x = input("Coordenada X: ")
        y = input("Coordenada Y: ")

        
        if not (x.isdigit() and y.isdigit()):
            print("Debes introducir números válidos.")
            continue

        x = int(x)
        y = int(y)

        contDisparos += 1
        lstTableroMac = dispararBarco(lstTableroMac, x - 1, y - 1, contDisparos)
        pintarTablero(lstTableroMac)

        if not quedanBarcos(lstTableroMac):
            print("¡Has hundido todos los barcos!")
            bolSeguir = False
        elif contDisparos >= 15:
            print("Has agotado tus disparos. Fin del juego.")
            bolSeguir = False
        else:
            strJugar = input("¿Quieres continuar jugando? (S/N): ")
            if strJugar.upper() == "S":
                bolSeguir = True
            else:
                bolSeguir = False

except Exception as e:
    print("Ha surgido un problema:", e)
```
